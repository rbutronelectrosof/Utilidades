```python
import requests
from bs4 import BeautifulSoup

def buscar_productos(url, termino_busqueda):
    # Hacer la petición a la página
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Encontrar los elementos que contienen los productos
    productos = soup.find_all('div', class_='producto')  # Ajusta esto según la estructura del sitio

    resultados = []
    for producto in productos:
        nombre = producto.find('span', class_='nombre').text
        precio = producto.find('span', class_='precio').text
        if termino_busqueda.lower() in nombre.lower():
            resultados.append({'nombre': nombre, 'precio': precio})

    return resultados

# Ejemplo de uso
url = 'https://ar.air-intra.com/2022/'
termino = '+3050'
resultados = buscar_productos(url, termino)
for r in resultados:
    print(f"Nombre: {r['nombre']}, Precio: {r['precio']}")
```

```python
import requests
from bs4 import BeautifulSoup

def obtener_datos_de_pagina_web(url):
    # Enviar una solicitud a la página web
    response = requests.get(url)

    # Verificar si la solicitud fue exitosa
    if response.status_code == 200:
        # Parsear el contenido de la solicitud con BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Ejemplo: Encontrar todos los párrafos en la página
        parrafos = soup.find_all('p')

        # Extraer y mostrar el texto de cada párrafo
        for para in parrafos:
            print(para.get_text())
    else:
        print(f"Error al recuperar los datos. Código de estado: {response.status_code}")

# Reemplaza con la URL de la página web que deseas extraer
url = 'https://www.invidcomputers.com/'
obtener_datos_de_pagina_web(url)
```
